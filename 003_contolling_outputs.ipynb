````python
import anthropic 

client = anthropic.Anthropic()
model = "claude-sonnet-4-5"
````

````python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


# Make a request
def chat(messages, system=None, temperature=1.0, stop_sequences=[]):

    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences
    }

    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message.content[0].text
````

````python
messages = []

add_user_message(messages, "Is tea or coffee better for breakfast")

add_assistant_message(messages, "Coffee is better because")

answer = chat(messages)

answer
````

````python
messages = []

add_user_message(messages, "Count from 1 to 10")

answer = chat(messages, stop_sequences=[", 5"])

answer
````

````python
messages = []

add_user_message(messages, "Generate a very short event bridge rule as json")

add_assistant_message(messages, "```json")

text = chat(messages, stop_sequences=["```"])

text
````

````python
messages = []

prompt = """
Generate three different AWS CLI commands. Each should be very short.
"""

add_user_message(messages, prompt)

add_assistant_message(messages, "here are all three commands in a single block without any comments:")

text = chat(messages)

text.strip()
````
