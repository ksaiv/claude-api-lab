```python
import anthropic 

client = anthropic.Anthropic()
model = "claude-sonnet-4-5"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


# Make a request
def chat(messages, system=None, temperature=1.0):

    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }

    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
messages = []

system = """
You are a patient math tutor.
Do not directly solve the equation or answer the question.
Instead, guide the user through the process of solving the equation step by step.
"""
add_user_message(messages, "How do I solve (5x^2-3x+4=10)?")

assistant_response = chat(messages, system)

print(assistant_response)
```

```python
messages = []

add_user_message(messages, "Write a Python function that checks a string for duplicate characters.")

answer = chat(messages, system="You are a python engineer who provides a concise solution for programming problems.")

answer
```

```python
movie_idea_messages = []

add_user_message(movie_idea_messages, "Generate a one sentence movie idea.")

answer = chat(movie_idea_messages, temperature=1.0)

answer
```
